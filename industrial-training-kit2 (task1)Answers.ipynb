# --------------------

```python
#Handling missing values in CouponCode 
df["CouponCode"] = df["CouponCode"].fillna("No Coupon")
```

```python
#Remove Duplicates
df = df.drop_duplicates()
df.duplicated
```

```python
#if we catch same Order ID but differ in other fields keep the First
df = df.drop_duplicates(subset=["OrderID"], keep="first")
```

# correct Data format (Date , numbers , text)

```python
#format all Date column 
df["Date"] = pd.to_datetime(df["Date"]).dt.strftime("%Y-%m-%d")
 
print(f"{df['Date'].head(10)}")
```

```python
#format all price columns 
for col in ["UnitPrice", "TotalPrice"]:
        df[col] = df[col].round(2)
```

```python
  print(f"{df['UnitPrice'].head()}")
  print(f"{df['TotalPrice'].head()}")
```

```python
#format all Text
text_cols = ["Product", "OrderStatus", "PaymentMethod", "ReferralSource","ShippingAddress", "CouponCode"]

for col in text_cols:
    df[col] = df[col].str.strip().str.title()
```

# Check Duplicates

```python
# Check Duplicate OrderIDs
final_id_dupes = df.duplicated(subset=["OrderID"]).sum()
id_check = "PASS" if final_id_dupes == 0 else "FAIL"
print(f"Duplicate OrderIDs  : {final_id_dupes}  → {id_check}")
```

```python
# Check  Incorrectly formatted dates
import re
final_bad_dates = df["Date"].apply(lambda x: not bool(re.fullmatch(r"\d{4}-\d{2}-\d{2}", str(x)))).sum()
date_check = "PASS" if final_bad_dates == 0 else "FAIL"
print(f"Bad date formats  : {final_bad_dates}  → {date_check}")
```
